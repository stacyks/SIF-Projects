Part 1: plotting closing prices for a single company over time period.
This was done for SBUX.
As seen in graph, closing prices rise for SBUX as days elapse since 2010-01-01

```python
##PART 1: plotting the closing prices for a single company over time period
#by: phoebe dainer
tickers = ["SBUX"]
start_date2 = datetime.datetime(2010, 1, 1)
end_date2 = datetime.datetime(2018, 1, 1)

#grabbing closing prices + their dates
closeSBUX = su.get_data('equity_raw', tickers, 
                        start_date=start_date2, end_date=end_date2,
                        attributes=['close'])

dates = closeSBUX.date #grabbing closing date list

#iterating through closing date list
#changing the values from strings into days elapsed between 2010-01-01 and said date.
daysElapsed = []
for date in dates:
    date = datetime.datetime.strptime(str(date), "%Y-%m-%d")
    time = date - start_date2
    time = time.days
    daysElapsed.append(int(time)) 

closeSBUX.date = daysElapsed #replacing the dates with its time elapsed since 2010-01-01

del closeSBUX['ticker'] #deleting ticker since useless

#plotting the closing prices over the time period 2010-2018
xpoints = (daysElapsed)
ypoints = (closeSBUX.close)

#for title
toString = ""
for ticker in tickers:
    toString = toString + ticker + " "

#plotting:
plt.scatter(xpoints, ypoints)
plt.title("Closing Prices of " + toString)
plt.xlabel('Days elapsed since 2010-01-01')
plt.ylabel('Closing prices in dollars')
```

Part 2: plotting 5 day rolling averages for a single company over a time period
This was done for SBUX, and done with closing prices.
As seen in graph, the rolling averages of closing prices generally rise as time elapses since 2010-01-01.
However, there are some points that deviate from the general shape of the graph. This may be due to large fluctuations in closing prices over the 5 days that are averaged.

```python
##PART 2: plotting 5 day rolling averages for a single company over a time period
#by: stacy sun
# change dates to datetime form, set to five day intervals
tickers = ["SBUX"]
start_date2 = datetime.datetime(2010, 1, 1)
end_date2 = datetime.datetime(2018, 1, 1)

#grabbing closing prices + their dates
closeSBUX = su.get_data('equity_raw', tickers, 
                        start_date=start_date2, end_date=end_date2,
                        attributes=['close'])

index = 1
allDates = []
for newDate in closeSBUX['date']:
    if index % 5 == 0:
        toAdd = datetime.datetime.strptime(str(newDate), "%Y-%m-%d")
        allDates.append(toAdd)
    index += 1

# create the dictionary with the new dates
rAvgDict = {
    'fiveDayPeriods': allDates
}

# add up average for five days/5 for rolling average
# add that to rAvgDict dictionary
rAvgList = []
index = 1

rollingTotal = 0.0

for closePrice in closeSBUX['close']: 
    rollingTotal += closePrice

    if index % 5 == 0:
        rollingTotal /= 5
        rAvgList.append(rollingTotal)
        rollingTotal = 0.0
        
    index +=1

rAvgDict.__setitem__('fiveDayRollingAvg',rAvgList)
rAvgdf = pd.DataFrame(rAvgDict)

#plotting the closing price rolling averages (5 day periods) 
xpoints = allDates
ypoints = rAvgList

toString = ""
for ticker in tickers:
    toString = toString + ticker + " "

plt.scatter(xpoints, ypoints)
plt.title("Rolling 5-Day Averages of Closing Prices of SBUX")
plt.xlabel('Year')
plt.ylabel('Closing Price Rolling Averages in Dollars')
```

```python
###EXTRA: another comparison of equity and fundamental data, but only for one company
#by: stacy sun
#calculate rolling averages of the months in the fundamental data. 
#only grab the ones for 3, 6, 10, 12.
#compare the investments in each month to the rolling average highs of the company for that month.  
tickers = ['COST']

equity_data = su.get_data('equity_raw', tickers, start_date = start_date2, end_date = end_date2, attributes = ['high'])
fundamental_data = su.get_data('fundamental_raw', tickers, start_date = start_date2, end_date = end_date2, attributes = ['investments'])

#convert fundamental date for better readability.
for index, highs in enumerate(fundamental_data['investments']):
    highs/= 10**8
    fundamental_data['investments'][index] = highs

#function to convert all dates to datetimes 
def convertAllDates (cat):
    allDates = []
    for newDate in cat:
            toAdd = datetime.datetime.strptime(str(newDate), "%Y-%m-%d")
            allDates.append(toAdd)
    return allDates

# sort the equity data. 
convertedDates = convertAllDates(equity_data['date'])
convertedDates.sort()

usedDates = [3, 6, 10, 12]
usedDatesIndex = 0
monthRAvg = 0.0
days = 0

RAvgForMonths = []

# for each date in the equity data, add all the relevant rolling averages to a list. 
for index, date in enumerate(convertedDates):
    month = date.month

    # if target month is found (3, 6, 10, 12), calculate the rolling average. 
    if (month == usedDates[usedDatesIndex]):
        monthRAvg += equity_data['high'][index]
        days += 1 
    elif (month == usedDates[(usedDatesIndex+1)%4]):
        # next month: append the rolling average to the list.
        RAvgForMonths.append(monthRAvg/days)
        usedDatesIndex = (usedDatesIndex+1) % 4
        monthRAvg = equity_data['high'][index]
        days = 1

RAvgForMonths.append(monthRAvg/days)


#fundamental data

# convert and sort the dates. 
fundamentalConvertedDates = convertAllDates (fundamental_data['datekey'])
fundamentalConvertedDates.sort()

usedDatesIndex = 0
monthInvestmentAvg = 0.0
monthInvAvgs = []

# append each piece of investment data for list to be used. 
for index, date in enumerate(fundamentalConvertedDates):
    monthInvAvgs.append(fundamental_data['investments'][index])


# create correlation graph. 
import matplotlib.pyplot as plt
#plotting the closing price rolling averages (5 day periods) 
xpoints = RAvgForMonths
ypoints = monthInvAvgs
# ypoints = monthInvAvgs

toString = ""
for ticker in tickers:
    toString = toString + ticker + " "

plt.scatter(xpoints, ypoints)
plt.title("Costco: Investments Made vs Rolling Month Averages")
plt.xlabel('Rolling Averages by Month')  
plt.ylabel('Investments Made')
```
